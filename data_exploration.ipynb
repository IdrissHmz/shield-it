```python
# # Total number of messages sent
# # Total number of messages received
# # Total number of messages sent at night
# # Total number of messages received at night
# # Total number of messages sent on holidays
# # Total number of messages received on holidays 
# # Total number of messages sent on weekends
# # Total number of messages received on weekend 
# # Avg_msg_len_received ++
# # Avg_msg_len_sent ++
# # Avg Ratio of 1st person to 3rd person pronoun of messages sent
# # Avg Ratio of 1st person to 3rd person pronoun of messages received
# # Average Number of Messages sent/day
# # Average Number of Messages received/day
# # Ratio Sent/Received
# # Last message sent datetime
# # Degree:


# Average delta_time_between_messages
# Constraint:
# Average emotions in messages
# Total volume of emails sent compared to a hierarchical level baseline.
# Total volume of emails received compared to a hierarchical level baseline.
# Higher Hierarchical Level Reciprocity: Number of superiors with whom there is a reciprocal communication
# Executive Hierarchical Level
# Staff Member Hierarchical Level
# influencability 
# Variance in the Number of Messages sent/day
# Variance in the Number of Messages received/day
```

```python
emails_req['to_recipients'] = emails_req.apply(lambda x: json.dumps([s.strip()[:4] if ('\n' in s) else s.strip() for s in x['to_recipients'].split(',')]),axis=1)
```

```python
empl_df = employee_df.copy()
empl_df['nb_mail_sent'] = 0
empl_df['nb_mail_received'] = 0
```

```python
grp_send = emails_req.groupby('from_email').agg('count')['sent_datetime']
grp_rec = emails_req.groupby('to_recipients').agg('count')['sent_datetime']
```

```python
def make_dictionary_sender(group):
    dictionay = {}
    for k,v in zip(group.index, group.values):
        dictionay[k]=v
    return dictionay

def make_dictionary_receiver(group):
    dictionay = {}
    grp_rec_val = list(group.values[1:])
    grp_rec_ind = list(group.index)

    grp_rec_ind = [g_list for g_list in grp_rec_ind if (not '<' in g_list) and g_list !='']
    
    for k,v in zip(grp_rec_ind, grp_rec_val):
        k = json.loads(k)
        if k[0]=='': continue
        if 1<len(k) :
            for f in k:
                dictionay[f] = v
        else:        
            dictionay[k[0]] = v
    return dictionay    
```

```python
nb_messages_sent_dict = make_dictionary_sender(grp_send)
nb_messages_rec_dict = make_dictionary_receiver(grp_rec)
```

```python
empl_df['nb_mail_sent'] = empl_df.apply(lambda x: nb_messages_sent_dict[x['email_address']] if x['email_address'] in nb_messages_sent_dict.keys() else 0, axis=1)
empl_df['nb_mail_received'] = empl_df.apply(lambda x: nb_messages_rec_dict[x['email_address']] if x['email_address'] in nb_messages_rec_dict.keys() else 0, axis=1)

# empl_df['nb_mail_sent_square'] = empl_df.apply(lambda x: nb_messages_sent_dict[x['email_address']]**2 if x['email_address'] in nb_messages_sent_dict.keys() else 0, axis=1)
# empl_df['nb_mail_received_square'] = empl_df.apply(lambda x: nb_messages_rec_dict[x['email_address']]**2 if x['email_address'] in nb_messages_rec_dict.keys() else 0, axis=1)
```

```python
grp_send_night = emails_req[emails_req['is_night']==True].groupby('from_email').agg('count')['sent_datetime']
grp_send_weekend = emails_req[emails_req['is_weekend']==True].groupby('from_email').agg('count')['sent_datetime']
grp_send_holyday = emails_req[emails_req['is_holiday']==True].groupby('from_email').agg('count')['sent_datetime']
grp_rec_night = emails_req[emails_req['is_night']==True].groupby('to_recipients').agg('count')['sent_datetime']
grp_rec_weekend  = emails_req[emails_req['is_weekend']==True].groupby('to_recipients').agg('count')['sent_datetime']
grp_rec_holyday = emails_req[emails_req['is_holiday']==True].groupby('to_recipients').agg('count')['sent_datetime']
```

```python
grp_send_night_dict = make_dictionary_sender(grp_send_night)
grp_send_weekend_dict = make_dictionary_sender(grp_send_weekend)
grp_send_holyday_dict = make_dictionary_sender(grp_send_holyday)

grp_rec_night_dict = make_dictionary_receiver(grp_rec_night)
grp_rec_weekend_dict = make_dictionary_receiver(grp_rec_weekend)
grp_rec_holyday_dict = make_dictionary_receiver(grp_rec_holyday)
```

```python
empl_df['nb_mail_sent_night'] = 0
empl_df['nb_mail_sent_weekend'] = 0
empl_df['nb_mail_sent_holiday'] = 0
empl_df['nb_mail_received_night'] = 0
empl_df['nb_mail_received_weekend'] = 0
empl_df['nb_mail_received_holiday'] = 0
```

```python
empl_df['nb_mail_sent_night'] = empl_df.apply(lambda x: grp_send_night_dict[x['email_address']] if x['email_address'] in grp_send_night_dict.keys() else 0, axis=1)
empl_df['nb_mail_sent_weekend'] = empl_df.apply(lambda x: grp_send_weekend_dict[x['email_address']] if x['email_address'] in grp_send_weekend_dict.keys() else 0, axis=1)
empl_df['nb_mail_sent_holiday'] = empl_df.apply(lambda x: grp_send_holyday_dict[x['email_address']] if x['email_address'] in grp_send_holyday_dict.keys() else 0, axis=1)

empl_df['nb_mail_received_night'] = empl_df.apply(lambda x: grp_rec_night_dict[x['email_address']] if x['email_address'] in grp_rec_night_dict.keys() else 0, axis=1)
empl_df['nb_mail_received_weekend'] = empl_df.apply(lambda x: grp_rec_weekend_dict[x['email_address']] if x['email_address'] in grp_rec_weekend_dict.keys() else 0, axis=1)
empl_df['nb_mail_received_holiday'] = empl_df.apply(lambda x: grp_rec_holyday_dict[x['email_address']] if x['email_address'] in grp_rec_holyday_dict.keys() else 0, axis=1)
```
