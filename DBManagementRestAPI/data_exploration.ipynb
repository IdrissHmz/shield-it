```python
data = pd.read_csv('/home/client/Documents/shield-it/DBManagementRestAPI/data/filtered_years_to_from.csv')

# Removing email content from first column
data = data.drop(columns=['Unnamed: 0', 'file', 'message_id', 'X-Folder', 'X-To'])


data["date"] = data["date"].astype("datetime64")
#data["hour"]=data["date"].dt.hour
```

```python
print(len(data['from_name'].unique()))
print(len(data['employee'].unique()))
print(len(data['from_email'].unique()))
```

```python
names = data['from_name'].unique()
dictio_name = {}
i = 0
for e in names:
    addr = data[data['from_name']==e]['from_email'].unique()
    if '<' in addr[0]:
        dictio_name[e]=addr[0][:addr[0].index('<')]
    else:
        dictio_name[e]=addr[0]   
    i+=1
    if i == 998: break     
```

```python
first_names = []
last_names = []
emails = []
for k,v in dictio_name.items():
    first_names.append(k.split()[0])
    last_names.append(k.split()[-1])
    emails.append(v)
```

```python
employee_df = pd.DataFrame({'first_name':first_names,'last_name':last_names,'email_address':emails})
employee_df
```
